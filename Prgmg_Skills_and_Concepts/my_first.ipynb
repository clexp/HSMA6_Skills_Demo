I would like to get some random data and plot it in a graph.  

```python
from random import randint as rn 
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

Now lets make a series with random numbers.  

```python
x, y = [ rn(0,100) for _ in range (60)],[  rn(50,400) for _ in range (60)]
plt.scatter(x,y)
plt.xlabel('jumbos')
plt.ylabel('mumbos')
plt.title('mumbos verses jumbos')
plt.show()
```

Now lets see if we can use a rust library

```python
import polars as pls
import random
import matplotlib.pyplot as plt
n = 50
x = [random.randint(0, 100) for _ in range(n)]
y = [random.randint(0, 100) for _ in range(n)]
data = pls.DataFrame({'x':x, 'y': y})
pdf = data.to_pandas()
plt.scatter(pdf['x'], pdf['y'])
plt.title('Random Scatter Plot')
plt.xlim([-5, 105])
plt.ylim([-5, 105])
plt.show()
```
